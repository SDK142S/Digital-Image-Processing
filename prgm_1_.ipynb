```python
#1. Simulation and Display of an Image, Negative of an Image(Binary & Gray Scale)

from PIL import Image, ImageOps
import matplotlib.pyplot as plt

def print_image(image_path):
    img = Image.open(image_path)
    negative_image = ImageOps.invert(img)
    gray_scale = ImageOps.grayscale(img)
    gray_scale_negative = ImageOps.invert(gray_scale)


    # Display original image
    plt.subplot(2, 2, 1)
    plt.imshow(img)
    plt.title("Original Image")
    plt.axis('off')


    # Display negative image
    plt.subplot(2, 2, 2)
    plt.imshow(negative_image)
    plt.title("Negative Image")
    plt.axis('off')

    # Display grayscale image
    plt.subplot(2, 2, 3)
    plt.imshow(gray_scale, cmap='gray')
    plt.title("Grayscale Image")
    plt.axis('off')

    # Display negative grayscale image
    plt.subplot(2, 2, 4)
    plt.imshow(gray_scale_negative, cmap='gray')
    plt.title("Negative Grayscale Image")
    plt.axis('off')

    plt.show()

image_path = "images/rose.jpeg"
print_image(image_path)

```
