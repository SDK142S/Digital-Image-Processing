```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Load images
im1 = Image.open("images/rose.jpeg")
im2 = Image.open("images/flower.jpg")

# Display original images
plt.figure(figsize=(12, 6))
plt.subplot(221), plt.title("First Image Before Resize"), plt.imshow(im1), plt.axis("off")
plt.subplot(222), plt.title("Second Image Before Resize"), plt.imshow(im2), plt.axis("off")

# Resize images
img1 = cv2.resize(np.array(im1), (im2.width, im2.height))
img2 = np.array(im2)

# Display resized images
plt.subplot(223), plt.title("First Image After Resize"), plt.imshow(img1), plt.axis("off")
plt.subplot(224), plt.title("Second Image After Resize"), plt.imshow(img2), plt.axis("off")

# Compute statistics
mean_img1, std_img1 = np.mean(img1), np.std(img1)
mean_img2, std_img2 = np.mean(img2), np.std(img2)
corr_coeff = np.corrcoef(img1.flatten(), img2.flatten())[0, 1]

# Display statistics
print("Mean of first image:", mean_img1)
print("Mean of second image:", mean_img2)
print("Standard Deviation of first image:", std_img1)
print("Standard Deviation of second image:", std_img2)
print("Correlation Coefficient:", corr_coeff)

plt.show()
```
