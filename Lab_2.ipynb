```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('images/forest.jpg', 0)
min_intensity = np.min(img)
max_intensity = np.max(img)

stretched_image = ((img - min_intensity) / (max_intensity - min_intensity) * 255).astype(np.uint8)
equalized_original = cv.equalizeHist(img)

hist_original = cv.calcHist([img], [0], None, [256], [0, 256])  # Use 256 bins instead of 255
hist_stretched = cv.calcHist([stretched_image], [0], None, [256], [0, 256])  # Use 256 bins
hist_equalized = cv.calcHist([equalized_original], [0], None, [256], [0, 256])  # Use 256 bins

plt.figure(figsize=(12, 8))

plt.subplot(231)
plt.title('Original Image')
plt.imshow(img, cmap='gray')  # Use cmap='gray' for grayscale images
plt.axis("off")

plt.subplot(232)
plt.title('Stretched Image')
plt.imshow(stretched_image, cmap='gray')  # Use cmap='gray' for grayscale images
plt.axis("off")

plt.subplot(233)
plt.title('Equalized Image')
plt.imshow(equalized_original, cmap='gray')  # Use cmap='gray' for grayscale images
plt.axis("off")

plt.subplot(234)
plt.title('Histogram - Original')
plt.bar(range(256), hist_original[:,0], color='r')
plt.xlim([0, 256])

plt.subplot(235)
plt.title('Histogram - Stretched')
plt.bar(range(256), hist_stretched[:,0], color='g')
plt.xlim([0, 256])

plt.subplot(236)
plt.title('Histogram - Equalized')
plt.bar(range(256), hist_equalized[:,0], color='b')
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```
