```python
#9) Implemetation of image Sharpening filters and Edege Detection using Gradient Filters
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Read the image
img = cv2.imread("images/rose.jpeg")

# Apply Gaussian Blur
blur_image = cv2.GaussianBlur(img, (3, 3), 0)

# Sharpening kernel
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])

# Apply Sharpening Filter
sharpen_image = cv2.filter2D(img, -1, kernel)

# Edge Detection using Canny
edges = cv2.Canny(image=blur_image, threshold1=100, threshold2=200)

# Display the results
plt.figure(figsize=(12, 6))

plt.subplot(221)
plt.title("Original Image")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(222)
plt.title("Blurred Image")
plt.imshow(cv2.cvtColor(blur_image, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(223)
plt.title("Sharpened Image")
plt.imshow(cv2.cvtColor(sharpen_image, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(224)
plt.title("Edge of Image")
plt.imshow(cv2.cvtColor(edges, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.tight_layout()
plt.show()
```
