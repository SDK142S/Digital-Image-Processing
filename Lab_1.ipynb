```python
#Read an image,display the image,negative and grayscale images

from PIL import Image, ImageOps
import matplotlib.pyplot as plt

def image_print(image_path):
    image = Image.open(image_path)
    
    negative_image = ImageOps.invert(image)
    
    grayscale_image = ImageOps.grayscale(image)
    
    grayscale_image_negative = ImageOps.invert(grayscale_image)
    
    plt.figure(figsize=(7, 5))
    
    plt.subplot(221)
    plt.title('Original Image')
    plt.imshow(image)
    plt.axis("off")
    
    plt.subplot(222)
    plt.title('Negative Original Image')
    plt.imshow(negative_image)
    plt.axis("off")
    
    plt.subplot(223)
    plt.title('Grayscale Image')
    plt.imshow(grayscale_image, cmap='gray')  # Use cmap='gray' for grayscale images
    plt.axis("off")
    
    plt.subplot(224)
    plt.title('Grayscale Negative Image')
    plt.imshow(grayscale_image_negative, cmap='gray')  # Use cmap='gray' for grayscale images
    plt.axis("off")
    
    plt.tight_layout()  # Ensures proper spacing between subplots
    plt.show()

image_print('images/flower.jpg')

```
