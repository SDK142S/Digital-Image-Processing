```python
#10) Implemetation of Image intensity slicing technique for image enhancement

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
img = cv2.imread("images/rose.jpeg", 0)

# Image intensity slicing parameters
min_range = 10
max_range = 60

# Apply intensity slicing
img1 = np.where((img > min_range) & (img < max_range), 255, 0).astype('uint8')

# Display the results
plt.figure(figsize=(12, 6))

plt.subplot(121)
plt.title("Original Image")
plt.imshow(img, cmap="gray")
plt.axis("off")

plt.subplot(122)
plt.title("Sliced Image")
plt.imshow(img1, cmap="gray")
plt.axis("off")

plt.tight_layout()
plt.show()
```
